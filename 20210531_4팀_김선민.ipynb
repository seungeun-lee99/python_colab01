```python
#예제. 편의점 재고관리 
#편의점 재고 관리 프로그램을 작성해보자.
#엔터를(while사용) 입력할때까지 반복해서 사용자로부터 물건 이름과 개수를 입력받는다. 
#만약 물건 이름이 존재하지 않으면 추가하고, 존재하면 물건 개수를 더한다(in사용 해서 존재 여부파악, if~else 구조 사용).
#입력이 끝나면 재고현황을 물건이름의 오름차순으로 정렬하여 출력한다(딕셔너리 전체를 물건 이름으로 정렬하여 for문을 돌려서 출력).
x={}
while True:
    product = input('물건이름: ') #재고 딕셔너리 
    if product =='':
        break
    number = int(input('개수: ')) #--> 숫자형이기 때문에 숫자로 입력해야함. 
    if product in x.keys(): #x딕셔너리 키가 product 안에 있으면 
        x[product] += number # +1을 해라 
    else:
        x[product] = number #아니면 추가해라 
print(x)
for key,value in sorted(x.items()): # 오름차순 정렬 
    print(key, value,end = ' ')
```

```python
#예제> 금액에 따른 할인율 계산하기 
#>물건 구매가를 입력받고, 금액에 따른 할인율을 계산하여 구매가, 할인율, 할인금액,지불금액을 출력하는 프로그램을 작성하시오.
p = int(input('물건 구매가를 입력하시오:'))
print('구매가:',p)
if p >= 100000:
    dc= .1
elif p >= 50000: 
    dc=0.075
else:
    dc=0.005
print(f'할인율: {dc*100} %')
print(f'할인금액: {p*dc}원' )
print(f'지불금액: {p-(dc*p)}원')
```

```python
#예제 최소값 구하기 
#3개의 수를 입력 받아 가장 작은 수를 출력하세요
#순저대로 정리가 안되어 있다면 if를 쓰면 안됨. 
n1 = int(input('첫번째 숫자를 입력하시오:'))
n2 = int(input('두번째 숫자를 입력하시오:'))
n3 = int(input('세번째 숫자를 입력하시오:'))
min1= n1# 첫번째 수 저장
if min1 > n2: # 두번째와 비교
    mini1 = n2      # 두번째수가 작으면 min1 저장

if n3 < min1:   # 세번째 수 비교
    min1 = n3# 세번째 수가 작으면 min1 저장


print('가장작은 수: ', min1)
```

```python
#예제> 사칙연산 프로그램 
#두 수와 사칙연산기호를 입력받아 연산기호에 따라 연산결과를 출력하는 프로그램을 작성하세요. 
n1 = int(input('첫번째 숫자를 입력하시오 :'))
n2 = int(input('두번째 숫자를 입력하시오 :'))
n3 = input('연산자를 입력하시오(+,-,*,/) :')
print('_'*50)
if n3 == '+':
    print(f' {n1} + {n2} = {n1+n2}')
elif n3 == ('-', 50):
    print(f' {n1} - {n2} = {n1-n2}')
elif n3 == '*':
    print(f' {n1} * {n2} = {n1*n2}')
elif n3 == '/' and n2 != 0:
    print(f' {n1} / {n2} = {n1/n2}')
else:
    print(f' {n1} / {n2} = 0으로 나눌 수 없습니다')
```
