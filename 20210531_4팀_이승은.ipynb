#예제3 > 편의점 재고 관리
엔터를 입력할 때까지 반복하여 사용자로부터 물건이름과 개수를 입력받는다.</br>
만약 물건 이름이 존재하지 않으면 추가하고, 존재하면 물건 개수를 더한다.</br>
입력이 끝나면 재고 현황을 물건이름의 오름차순으로 정렬하여 출력한다.

```python
a={}

while True:
    product = input('물건이름:')
    if product=='':
        break
    number = int(input('개수:'))
    if product in a.keys():
        a[product] += number
    else:
        a[product] = number
print(a)

#오름차순 정렬(0~9, ㄱ~ㅎ 등)
for key,value in sorted(a.items()):
    print(key,value, end=' ')
```

#집합(set) > 연산
a와 b의 합집합 a.union(b) -> a|b</br>
a와 b의 교집합 a. intersection(b) -> a&b</br>
a와 b의 차집합 a.difference(b) -> a-b</br>


```python
sweety = {'빼빼로', '칸초', '허니버터'}
salty = {'허니버터', '포카칩', '새우깡'}

#합집합
print(sweety.union(salty))
print(sweety|salty)

#교집합
print(sweety.intersection(salty))
print(sweety&salty)

#차집합
print(sweety.difference(salty))
print(sweety-salty)
```

```python
#요소 추가
#집합명.add(요소)       #1개 추가
set_wish = {'새우깡','오감자','양파링','포카칩','양파링'}
set_wish.add('허니버터')
print(set_wish)

#집합명.update(요소)    #여러개 추가
set_wish1 = {'새우깡','오감자','양파링','포카칩','양파링'}
set_wish1.update(['빼빼로','칸초'])
print(set_wish1)

set_wish1 = {'새우깡','오감자','양파링','포카칩','양파링'}
set_wish1.update(['빼빼로'])
print(set_wish1)
set_wish1.update('빼빼로')     #[]를 사용하지 않으면 각 문자가 하나씩 추가됨
print(set_wish1)
```

```python
#요소 삭제
#집합명.remove(요소) -> 인덱스가 없기 때문에  remove만 사용 가능
set_wish = {'새우깡','오감자','양파링','포카칩','양파링'}
set_wish.remove('오감자')
print(set_wish)

#빼빼로 삭제
#존재하지 않는 요소 삭제 시도 -> 오류 발생
set_wish1 = {'새우깡','오감자','양파링','포카칩','양파링'}
if '빼빼로' in list(set_wish1):               #집합을 리스트로 만들어서 요소 확인해야 함!
    set_wish1.remove('빼빼로')
print(set_wish1)
```

```python
#예제 > 금액에 따른 할인율 계산하기
#물건 구매가를 입력받고, 금액에 따른 할인율을 계산하여 구매가, 할인율, 할인금액, 지불금액을 출력하는 프로그램을 작성하시오.

p = int(input('물건 구매가를 입력하세요:'))
if p >= 100000:
    dc = 10
    price = p * (1-dc/100)
elif p >= 50000:
    dc = 7.5
    price = p* (1-dc/100)
elif p >= 10000:
    dc = 5
    price = p* (1-dc/100)
else:
    dc = 0
    price = p

print(f'구매가:{p}원')
print(f'할인율:{dc}%')
print(f'할인금액:{int(p-price)}원')
print(f'지불금액:{int(price)}원')
```

#논리연산자
논리 연산자를 사용하여 여러개의 조건을 조합하여 복잡한 조건식을 만들 수 있다.</br>
and : 그리고 : 모두 True 일 때만 True</br>
or : 또는 : 하나 이상이 True이면 True</br>
not : ~가 아니다 : 결과를 반대로 출력 -> False 일때만 실행(True) : 거짓이면 참이다.</br>

```python
# 논리 연산자(Logical operator)
# and, or, not
print(True and True)
print(True and False)
print(False and True)
print(False and False)
print()
print(True or True)
print(True or False)
print(False or True)
print(False or False)
print()
print(not True)
print(not False)         #거짓일때만 참이다.
```

```python
#예제 > 사칙연산 프로그램
#두 수와 사칙연산기호(+,-,*,/)를 입력받아 연산 기호에 따라 연산 결과를 출력하는 프로그램 작성
n1=int(input('숫자1:'))
n2=int(input('숫자2:'))
i=input('연산기호(+,-,*,/):')

if i == '+':
    print(f'{n1} {i} {n2} = {n1 + n2}')
elif i == '-':
    print(f'{n1} {i} {n2} = {n1 - n2}')
elif i == '*':
    print(f'{n1} {i} {n2} = {n1 * n2}')
elif i == '/':
    if n2 == 0:
        print('계산 불가')
    else:
        print(f'{n1} {i} {n2} = {int(n1 / n2)}')
```

```python
#예제 > 주사위 던지기 게임
#두 개의 주사위를 던져 더 큰 수가 나오는 사람이 이기는 게임을 작성하시오.
import random
d1=random.randint(1,6)
d2=random.randint(1,6)
print('주사위1:',d1)
print('주사위2:',d2)

if d1 > d2:
    print('주사위1이 이겼습니다.')
elif d1 < d2:
    print('주사위2가 이겼습니다.')
elif d1 == d2:
    print('주사위1과 주사위2가 동점입니다.')
```
